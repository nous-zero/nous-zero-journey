# LeetCode #217 - Contains Duplicate

**날짜**: 2026-04-16  
**난이도**: Easy  
**주제**: Array / Hashing

## 문제

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

정수 배열 nums가 주어졌을 때, 배열에 어떤 값이 두 번 이상 나타나면 true를 출력하고, 모든 요소가 서로 다르면 false를 출력하세요.

### Examples
- Input: `nums = [1,2,3,1]` → Output: `true` (1이 index 0, 3에 두 번 등장)
- Input: `nums = [1,2,3,4]` → Output: `false` (모두 다름)
- Input: `nums = [1,1,1,3,3,4,3,2,4,2]` → Output: `true`

### Constraints
- 1 <= nums.length <= 10^5
- -10^9 <= nums[i] <= 10^9

## 풀이 1: 이중 for 문 (Brute Force)

- 시간복잡도: O(n²)
- 공간복잡도: O(1)
- 배열이 크면 느림 (nums.length가 10^5이면 10^10번 연산)

````python
from typing import List

class SolutionForLoop:
    def containsDuplicate(self, nums: List[int]) -> bool:  # 중복성에 대한 기능
        for i in range(len(nums)):                    # 첫번째 숫자
            for j in range(i + 1, len(nums)):         # 첫번째 숫자 다음 숫자들과 비교
                if nums[i] == nums[j]:                # i와 j의 같은 숫자 있다면
                    return True                       # True를 출력
        return False                                  # 없으면 False 출력

# 테스트
sol = SolutionForLoop()
print(sol.containsDuplicate([1, 2, 3, 1]))    # True
print(sol.containsDuplicate([1, 2, 3, 4]))    # False
````

## 풀이 2: 집합(Set) 사용

- 시간복잡도: O(n)
- 공간복잡도: O(n)
- `set(nums)`은 중복을 자동으로 제거 → 길이 비교로 중복 여부 판단

````python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        return len(nums) != len(set(nums))
        # 중복을 자동으로 제거한 집합이 원래 길이와 다르면 중복 있음

# 테스트
sol = Solution()
print(sol.containsDuplicate([1, 2, 3, 1]))                  # True
print(sol.containsDuplicate([1, 2, 3, 4]))                  # False
print(sol.containsDuplicate([1, 1, 1, 3, 3, 4, 3, 2, 4, 2]))  # True
````

## 학습 정리: for 문 / set

### for 문 기본 구조

```
for 변수 in 반복할것:
    # 반복할것 안에서 하나씩 꺼내서 -> 변수에 담고 -> 코드 실행
```

### 예시

```python
# range(5) -> 0, 1, 2, 3, 4
for i in range(5):
    print(i)

# range(2, 6) -> 2, 3, 4, 5
for i in range(2, 6):
    print(i)

# 리스트 순회
nums = [10, 20, 30]
for n in nums:
    print(n)  # 10, 20, 30

# 중첩 for 문
for i in range(3):
    for j in range(3):
        print(i, j)  # (0,0), (0,1), ... (2,2)
```

## 회고

- 첫 번째 풀이(이중 for)는 직관적이지만 O(n²)로 느림
- 두 번째 풀이(set)는 Python의 내장 자료구조로 O(n) 해결
- **교훈**: 집합의 '중복 제거' 특성을 이용하면 한 줄로 해결 가능
